### Loading Cleaned Data and Training the Random Forest Model

```python
# Loading required libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
```

```python
# Loading data
df = pd.read_csv("data/titanic_clean.csv")
# Select relevant features
features = ['Pclass', 'Sex', 'Age', 'Fare', 'FamilySize', 'IsAlone']
df['Sex'] = df['Sex'].map({'male': 1, 'female':0})
target = 'Survived'
X = df[features]
y = df[target]
# Splitting data into train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3, test_size = 0.2)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
# training random forest model
model = RandomForestClassifier(random_state = 42)
model.fit(X_train, y_train)
```

```python
# making predictions and calculating accuracy on test data
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Random Forest Model Accuracy: {acc:.4f}")
```

```python
# ROC AUC 
auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
print(f"AUC: {auc:.3f}")
```

### Model Robustness Testing

```python
# Adding noise in data
np.random.seed(43)
X_noisy = X_test.copy()
X_noisy['Age'] = X_noisy['Age'] + np.random.normal(0, 0.2, size = len(X_noisy))
X_noisy['Fare'] = X_noisy['Fare'] * np.random.uniform(0.9, 1.1, size = len(X_noisy))

# making predictions on noisy data
y_pred_noisy = model.predict(X_noisy)
acc_noisy = accuracy_score(y_test, y_pred_noisy)
print(f"Random Forest Model Accuracy (Nosiy Data): {acc_noisy:.4f}")
stability = (y_pred == y_pred_noisy).mean()
print(f"Prediction Stability after Noise Injection: {stability:.3f}")
```

To evaluate how well the Random Forest model handles data variability, controlled random noise was injected into the numeric features (`Age` and `Fare`). The model was then re-evaluated on this perturbed dataset to measure its resilience.

|---------|---------|
| **Model** | Random Forest |
| **Accuracy (Noisy Data)** | 0.7709 |
| **Prediction Stability** | 0.950 |

**Interpretation:**

- The **accuracy decreased slightly to 0.7709**, indicating a minor sensitivity to input fluctuations but overall strong generalization capability.

**Conclusion:**  
The Random Forest classifier exhibits **high robustness** â€” maintaining strong performance and stable predictions even under noisy conditions.  
In practical applications, this level of stability implies the model would continue to perform consistently across imperfect or slightly varied input data.

### Advanced Feature Explainability
### 1. Permutation Importance

```python
from sklearn.inspection import permutation_importance
perm = permutation_importance(model, X_test, y_test, n_repeats=20, random_state=42)
imp = pd.DataFrame({'Feature': X_test.columns, 'Importance': perm.importances_mean}).sort_values(by = 'Importance', ascending = False)
import seaborn as sns
import matplotlib.pyplot as plt

# Set professional theme
sns.set_theme(style="whitegrid", font_scale=1.2)

# Sort by importance for clear visualization
imp_sorted = imp.sort_values('Importance', ascending=True)

# Create figure
plt.figure(figsize=(10, 6))

# Plot with a modern palette and clean edges
ax = sns.barplot(
    x='Importance', 
    y='Feature', 
    data=imp_sorted, 
    palette='viridis', 
    edgecolor='black'
)

# Add title and axis labels
plt.title('Permutation Feature Importance', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Mean Importance Score', fontsize=13)
plt.ylabel('Feature Name', fontsize=13)

# Add value labels on bars
for i in ax.containers:
    ax.bar_label(i, fmt='%.3f', label_type='edge', fontsize=10, padding=2)

# Improve layout
sns.despine(left=True, bottom=True)
plt.tight_layout()

# Display
plt.show()
```

**Interpretation Summary:**

**Interpretation Summary:**

- **Age and Fare** display clear, interpretable patterns aligned with real-world logic: youth and wealth increased survival likelihood.

- The model captures **non-linear relationships** effectively, demonstrating that the Random Forest algorithm leverages both hierarchical and interaction-based effects in prediction.

**Conclusion:**

### Fairness & Bias Analysis

### 1. Grouped Performance Check

```python
def group_metrics(df, group_col):
    metrics = []
    for val, subset in df.groupby(group_col):
        acc = accuracy_score(subset['Survived'], subset['Pred'])
        prec = precision_score(subset['Survived'], subset['Pred'])
        rec = recall_score(subset['Survived'], subset['Pred'])
        metrics.append([val, acc, prec, rec])
    return pd.DataFrame(metrics, columns=[group_col, 'Accuracy', 'Precision', 'Recall'])

df_eval = X_test.copy()
df_eval['Survived'] = y_test
df_eval['Pred'] = model.predict(X_test)

bias_table = group_metrics(df_eval, 'Sex')
print(bias_table)
```

### âš–ï¸ Grouped Performance Analysis (Fairness Check by Gender)

To evaluate potential **bias and fairness** in the modelâ€™s predictions, performance metrics were calculated separately for male (`Sex = 0`) and female (`Sex = 1`) passengers.

| Sex | Accuracy | Precision | Recall |
|-----|-----------:|-----------:|-----------:|
| **0 (Male)** | 0.770 | 0.800 | 0.906 |
| **1 (Female)** | 0.819 | 0.417 | 0.294 |

**Interpretation:**

- The **model performs well for male passengers**, achieving high precision (0.80) and very strong recall (0.91).  
  This means the model correctly identifies most males who did not survive, but potentially at the cost of some false negatives.  
- For **female passengers**, performance drops sharply â€” **precision (0.42)** and **recall (0.29)** are much lower.  
  This suggests the model **underestimates female survival probabilities**, failing to capture certain gender-specific survival patterns.  
- While the overall accuracy difference (0.77 vs. 0.82) is modest, the **recall imbalance** indicates **potential bias** â€” the model favors predicting â€œnon-survivalâ€ across the board, especially for females.

**Possible Causes:**
- **Data imbalance:** More males than females in the dataset could lead the model to learn a male-dominant survival pattern.  
- **Feature interaction:** Features like `Pclass`, `Fare`, and `FamilySize` may correlate differently with survival for each gender, which a simple model cannot fully disentangle.  
- **Threshold effect:** A single decision threshold (0.5) might not be optimal for both groups â€” separate calibration could improve fairness.

**Conclusion:**
The model exhibits **gender-based performance disparity**, performing considerably better for males.  
Addressing this may involve:
- Rebalancing or stratified sampling during training,  
- Introducing **fairness-aware loss functions**, or  
- Adjusting decision thresholds per subgroup.  

This step highlights the importance of **ethical model evaluation** â€” ensuring models are not only accurate but also **equitable across demographic groups.**
